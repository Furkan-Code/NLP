```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv("/content/Kitap3.csv")
df.head()
```

```python
df["category"].value_counts()
```

```python
x=df["text"]
y=df["category"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2)
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
import torch,sklearn
```

```python
model_args = {
    "use_early_stopping": True,
    "early_stopping_delta": 0.01,
    "early_stopping_metric": "mcc",
    "early_stopping_metric_minimize": False,
    "early_stopping_patience": 5,
    "evaluate_during_training_steps": 1000,
    "fp16": False,
    "num_train_epochs":2
}
```

```python
cuda_available = torch.cuda.is_available()
model = ClassificationModel(
    "bert",
    "dbmdz/bert-base-turkish-cased",
     use_cuda=cuda_available,
     args=model_args,
     num_labels=5
)
```

```python
train_df = pd.DataFrame({"text":x_train, "category":y_train})
eval_df = pd.DataFrame({"text":x_test,"category":y_test})
train_df.columns = ["category", "text"]
train_df.head()
from sklearn.preprocessing import LabelEncoder

# Etiketleri sayısallaştırma
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Yeniden DataFrame oluşturma
train_df = pd.DataFrame({"text": x_train, "category": y_train_encoded})

train_df.head()
```

```python
model.train_model(train_df, acc=accuracy_score)
```

```python
from flask import Flask, request, jsonify
from transformers import pipeline, AutoModelForSequenceClassification, AutoTokenizer

app = Flask(__name__)

# Model ve tokenizer yükleniyor
tokenizer = AutoTokenizer.from_pretrained("/content/outputs")
model = AutoModelForSequenceClassification.from_pretrained("/content/outputs")
nlp = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

@app.route('/predict', methods=['POST'])
def predict():
    data = request.get_json()
    text = data['text']
    # Metni analiz edip sonucu döndür
    result = analyze_text(text)
    return jsonify({'prediction': result})

def analyze_text(text):
    # NLP modelini kullanarak metni analiz edin
    analysis_result = nlp(text)
    # 'label' içindeki değere göre 'yes' veya 'no' dön
    if analysis_result[0]['label'] == 'LABEL_1':  # LABEL_1 pozitif duyguyu temsil ediyorsa
        return "yes"
    else:
        return "no"

if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0')
```
